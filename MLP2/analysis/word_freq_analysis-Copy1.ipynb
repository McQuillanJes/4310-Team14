```python
import matplotlib.pyplot as plt
import pandas as pd

# read data from CSV file to dataframe
words = pd.read_csv('./unigram_freq.csv')
answers = pd.read_csv('./answers.csv')
```

```python
# make sure you understand the type of the object
print(type(words))
print(type(answers))
```

```python
# check the top five and the bottom five data tuples
display(words.head())
display(words.tail())
display(words.shape)

display(answers.head())
display(answers.tail())
display(answers.shape)
```

```python
df = answers.join(words.set_index('word'), on='word')
display(df.head())
display(df.tail())
display(df.describe())
df.info()
```

```python
#Normalizing the count in the word
df['Normalized_count'] = (df['count'] - df['count'].min()) / (df['count'].max() - df['count'].min())

#Printing the head
display(df.head())
```

```python
#Sortting the values by normalized count
df = df.sort_values(by=['Normalized_count'], ascending = False)

#Printing the head
display(df.head())
```
